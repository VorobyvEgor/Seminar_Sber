### Задание 1

Написать функцию, которая принимает на вход список, на выход отдает список такого же размера. Выходной список формируется следующим образом: на i месте стоит произведение всех элементов ненулевых элементов входного списка кроме i. Необходимо это сделать как можно оптимальнее (в идеале за 2 прохода по массиву).

Ввод: `[1, 2, 3]`

Вывод: `[6, 3, 2]`

Пояснение: `2 * 3 = 6`, `1 * 3 = 3`, `1 * 2 = 2`

```python
x = [int(i) for i in input().split()]
result = [None for i in range(len(x))]
for i in range(len(x)):
  y = x.copy()
  y.remove(x[i])
  dev = 1
  if y.count(0) < len(y):
    for j in y:
      if j != 0:
        dev *= j
    result[i] = dev
print(result)
```
